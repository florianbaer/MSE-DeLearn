```python
import numpy as np
import torch
import matplotlib
import matplotlib.cm as cm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
```

### Utilities

The test function that is used for illustrating the workings of the optimisers.

```python
def test_function(x):
    return x[0]**2/15.0+x[1]**2
```

Utility for plotting the contours of a given function`f` in a range `r` (4-tuple) and 20 levels up to a max level given by `s`. 

```python
def plot_contours(r, s, f):
    matplotlib.rcParams['xtick.direction'] = 'out'
    matplotlib.rcParams['ytick.direction'] = 'out'
    delta = 0.01
    X = np.arange(r[0], r[1], delta)
    Y = np.arange(r[2], r[3], delta)
    X,Y = np.meshgrid(X, Y)  
    XX = np.stack((X,Y), axis=0)
    Z=f(XX)
    V=np.arange(0.2,s,s/20)**2
    #CS = plt.contour(X, Y, Z, V, colors=colors)
    #plt.clabel(CS, inline=1, fontsize=10)
    CS = plt.contour(X, Y, Z, V)
```

Optimisation (i.e. minimisation) of the function `f` starting at `x` (pytorch tensor), by using the given `optimizer` over`nepochs`epochs.  

```python
def optimize(f, x, nepochs, optimizer):
    params = [x,]
    xhist = [x[0].item()]
    yhist = [x[1].item()]
    optimizer.param_groups[0]['params'] = params
    for i in range(nepochs):
        y = f(x)
        optimizer.zero_grad()
        y.backward()
        optimizer.step()
        xhist.append(x[0].item())    
        yhist.append(x[1].item())
    return xhist, yhist
```

RMS error for measuring the discrepancy to the final target (for the test function above it is (0,0). 

```python
def error(xhist,yhist):
    x,y = xhist[-1], yhist[-1]
    return "%6.3f, (%6.3f, %6.3f)"%(np.sqrt(x**2+y**2),x,y)
```

### Sample Run

```python
nepochs = 20

plt.figure(1,figsize=(12,6))
plot_contours([-3.5,0.,-4.,4.], 4., test_function)

# initial value
x = torch.tensor([-3.0,-4.0]).requires_grad_()

optimizer = torch.optim.SGD([x],lr=.3)
xhist, yhist = optimize(test_function, x, nepochs, optimizer)  

plt.plot(xhist,yhist,"b-+", label="lr=0.3")
print("SGD - lr=0.3 ", error(xhist,yhist))
```

* We did so below
* Yes we could verify this with a learning rate $\alpha$ of 1

```python
def plot_learning(lr=0.3, nepochs=20):

    plt.figure(1,figsize=(12,6))
    plot_contours([-3.5,0.,-4.,4.], 4., test_function)

    # initial value
    x = torch.tensor([-3.0,-4.0]).requires_grad_()

    optimizer = torch.optim.SGD([x],lr=lr)
    xhist, yhist = optimize(test_function, x, nepochs, optimizer)  
    x = torch.tensor([-3.0,-4.0]).requires_grad_()

    plt.plot(xhist,yhist,"b-+", label=f"lr={lr}")
    print(f"SGD - lr={lr} ", error(xhist,yhist))
    
def plot_learning_momentum(lr=0.3, nepochs=20, momentum=0, nesterov=False):

    plt.figure(1,figsize=(12,6))
    plot_contours([-3.5,0.,-4.,4.], 4., test_function)

    # initial value
    x = torch.tensor([-3.0,-4.0]).requires_grad_()

    optimizer = torch.optim.SGD([x],lr=lr, momentum=momentum, nesterov=nesterov)
    xhist, yhist = optimize(test_function, x, nepochs, optimizer)  
    x = torch.tensor([-3.0,-4.0]).requires_grad_()

    plt.plot(xhist,yhist,"r-+", label=f"lr={lr}")
    print(f"Momentum - lr={lr} ", error(xhist,yhist))
    
def plot_learning_rms(lr=0.3, nepochs=20, alpha=0.99):

    plt.figure(1,figsize=(12,6))
    plot_contours([-3.5,0.,-4.,4.], 4., test_function)

    # initial value
    x = torch.tensor([-3.0,-4.0]).requires_grad_()

    optimizer = torch.optim.RMSprop([x],lr=lr, alpha=alpha)
    xhist, yhist = optimize(test_function, x, nepochs, optimizer)  
    x = torch.tensor([-3.0,-4.0]).requires_grad_()

    plt.plot(xhist,yhist,"g-+", label=f"lr={lr}")
    print(f"RMS - lr={lr} ", error(xhist,yhist))
    
def plot_learning_adam(lr=0.3, nepochs=20, betas=(0.9, 0.999)):

    plt.figure(1,figsize=(12,6))
    plot_contours([-3.5,0.,-4.,4.], 4., test_function)

    # initial value
    x = torch.tensor([-3.0,-4.0]).requires_grad_()

    optimizer = torch.optim.Adam([x], lr=lr, betas=betas)
    xhist, yhist = optimize(test_function, x, nepochs, optimizer)  
    x = torch.tensor([-3.0,-4.0]).requires_grad_()

    plt.plot(xhist,yhist,"k-+", label=f"lr={lr}")
    print(f"ADAM - lr={lr} ", error(xhist,yhist))
    
def plot_learning_adagrad(lr=0.3, nepochs=20):

    plt.figure(1,figsize=(12,6))
    plot_contours([-3.5,0.,-4.,4.], 4., test_function)

    # initial value
    x = torch.tensor([-3.0,-4.0]).requires_grad_()

    optimizer = torch.optim.Adagrad([x], lr)
    xhist, yhist = optimize(test_function, x, nepochs, optimizer)  
    x = torch.tensor([-3.0,-4.0]).requires_grad_()

    plt.plot(xhist,yhist,"c-+", label=f"lr={lr}")
    print(f"AdaGrad - lr={lr} ", error(xhist,yhist))
```
