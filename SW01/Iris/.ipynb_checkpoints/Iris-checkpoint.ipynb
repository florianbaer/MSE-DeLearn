```python
import torch, matplotlib.pyplot as plt, numpy as np, pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('iris.txt', sep='\t')
df.head()
```

```python
df.Species.unique()
```

```python
colors = {'setosa':'red', 'virginica':'blue', 'versicolor':'green'}
indexer = 1

plt.figure(figsize=(16,16))

for y in ['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']:
    for x in ['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']:
        plt.subplot(4, 4, indexer)
        if x == y:
            plt.hist(df[x])
            plt.xlabel(x)
        else:
            plt.scatter(df[x], df[y], c=df['Species'].map(colors))
            plt.xlabel(x)
            plt.ylabel(y)
        indexer = indexer + 1
plt.show()
```

It would be easier to separate the _setosa_ from the other two species.
The other two could "easy" be separated by the _petal length_ and the _petal width_ with a line separating these two kind of species.
